```python
class settings:
    def __init__(self, **kwargs):
        self.plc_name = kwargs.get('plc_name', 'unknown')
        self.ip = kwargs.get('ip', 'unknown')
        self.port = kwargs.get('port', 'unknown')
        self.db = kwargs.get('db', 'unknown')

    def show_settings(self):
        print(
            f'''
            PLC Name: {self.plc_name}
            IP Address: {self.ip}
            Port: {self.port}
            DB: {self.db}
            '''
        )

    def update_settings(self):
        print(
            f'''
            PLC Name: {self.plc_name}
            IP Address: {self.ip}
            Port: {self.port}
            DB: {self.db}
            '''
        )
```

```python
class model:
    def __init__(self):
        self.fields = {
            f'field_{i}': [random.uniform(0, 100) for _ in range(100)] for i in range(20)
        }

        self.filter = {
            'gt': self.gt,
            'lt': self.lt,
            'gte': self.gte,
            'lte': self.lte,
            'eq': self.eq,
        }

    def get_data(self, **kwargs):
        for key_suf, value in kwargs.items():
            parts = key_suf.split('__')
            key = parts[0]
            suf = parts[1]
            print(f'{key}: {self.filter[suf](key, value)}')



    def show_data(self):
        for field, value in self.fields.items():
            print(f'{field}: {value}')

    def gt(self, field, value):
        return [data for data in self.fields[field] if data > value]

    def lt(self, field, value):
        return [data for data in self.fields[field] if data < value]
    
    def gte(self, field, value):
        return [data for data in self.fields[field] if data >= value]

    def lte(self, field, value):
        return [data for data in self.fields[field] if data <= value]
    
    def eq(self, field, value):
        return [data for data in self.fields[field] if data <= value]
```

```python
model_1 = model()
```

```python
model_1.show_data()
```
