### aggregate: avg, sum, max, min, count

```python
from django.db.models import Avg, Sum, Max, Min, Count, StdDev, Variance
```

```python
metrics = Measurement.objects.aggregate(Avg('pressure'), Max('pressure'), Variance('pressure'), StdDev('pressure'))
metrics
```

```python
temperature = Measurement.objects.filter(temperature__gte=28).order_by('timestamp').first()
temperature
```

```python
temperature = Measurement.objects.filter(temperature__gte=28).order_by('timestamp').last()
temperature
```

```python
temperature = Measurement.objects.filter(temperature__gte=28).order_by('timestamp')
temperature
```

```python
Measurement.objects.filter(temperature__gte=28).order_by('timestamp').reverse()[0]
```
