```python
from datetime import datetime, timedelta
from django.db import DatabaseError
from time import sleep
import django
import pytz
import os

os.environ.setdefault("DJANGO_SETTINGS_MODULE", "industrial_insights.settings")
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true" # jupyter
django.setup()
```

```python
from home.models import Measurement
```

```python
colombia_tz = pytz.timezone("America/Bogota")
```

```python
import random
```

```python
for _ in range(100):
    object = Measurement.objects.create(
        timestamp = datetime.now(colombia_tz),
        pressure = round(random.uniform(0.1, 20.0), 2),
        temperature = round(random.uniform(20.0, 30.0), 2),
        box_units = int(random.uniform(0, 1000)),
        kwh = round(random.uniform(1.0, 2.0), 2),
        flow = round(random.uniform(50.0, 100.0), 2)
    )
    print(object)
    sleep(30)
```
